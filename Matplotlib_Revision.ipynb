# **Histograms**

```python
ages = np.random.normal(20,1.5, 1000)         #(mean, std, 1000)
```

```python
plt.hist(ages, bins=20, edgecolor="yellow", lw=2)
plt.show()
```

```python
plt.hist(ages, bins=20, cumulative=True)
plt.show()
```

# **Pie Charts**

```python
languages=['C++','JavaScript','Python','Go','PHP']
votes=[50,75,150,25,10]
```

```python
explodes=[0,0,0,0,0.3]
```

```python
plt.pie(votes, labels=languages, autopct='%1.1f%%', explode=explodes, startangle=90)
plt.show()
```

# **Customization**

```python
years = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]
income = [55, 56, 62, 61, 72, 72, 73, 75]
```

```python
income_ticks = list(range(50,81,2))
```

```python
plt.plot(years, income)
plt.title('Income')
plt.xlabel('Years')
plt.ylabel('Income')
plt.yticks(income_ticks, [f"{x}k USD" for x in income_ticks])
plt.show()
```

# **Multiple Figures**

```python
x1, y1 = np.random.random(100), np.random.random(100)
x2, y2 = np.arange(100), np.random.random(100)
```

```python
plt.figure(1)
plt.scatter(x1,y1)

plt.figure(2)
plt.plot(x2,y2)

plt.show()
```

```python
x = np.arange(100)

fig, axs = plt.subplots(2,2)

axs[0,0].plot(x, np.sin(x))
axs[0,0].set_title('Sine Wave')

axs[0,1].plot(x, np.cos(x))
axs[0,1].set_title('Cosine Wave')

axs[1,0].plot(x, np.random.random(100))
axs[1,0].set_title('Random Function')

axs[1,1].plot(x, np.log(x))
axs[1,1].set_title('Log Function')

fig.suptitle('Four Plots')
plt.show()
```
